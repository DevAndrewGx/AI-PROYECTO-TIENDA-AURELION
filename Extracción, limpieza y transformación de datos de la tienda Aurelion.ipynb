Queremos verificar si las columnas precio e importe necesitan ser transformadas al tipo de dato float.

```python
Buscamos_datos_tipo_float_en_precios = [x for x in productos_df['precio_unitario'].unique().tolist() if isinstance(x, float)]
print(Buscamos_datos_tipo_float_en_precios)
```

```python
Buscamos_datos_tipo_float_en_precio_unitario_DV = [x for x in detalle_ventas_df['precio_unitario'].unique().tolist() if isinstance(x, float)]
print(Buscamos_datos_tipo_float_en_precio_unitario_DV)
```

```python
Buscamos_datos_tipo_float_en_importe = [x for x in detalle_ventas_df['importe'].unique().tolist() if isinstance(x, float)]
print(Buscamos_datos_tipo_float_en_importe)
```

Por lo tanto, no hay datos tipo float en las columnas 'precio_unitario' y 'importe'.
